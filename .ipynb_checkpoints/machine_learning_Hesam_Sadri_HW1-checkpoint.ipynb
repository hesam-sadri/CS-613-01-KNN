```python
## Importing Libraries
import numpy as np
from sklearn import neighbors
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.neighbors import DistanceMetric
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import csv
```

## Exercise 1

```python
with open('cars.data.txt', 'r',) as csvfile:
    lines = csv.reader(csvfile)
    dataset = np.array(list(lines))
    dataset = dataset[1:,:]
```

```python
dataset.shape
```

```python
labels = dataset[:,3]
X = dataset[:,:-1].astype(np.float64)
```

```python
le = preprocessing.LabelEncoder()
labels = le.fit_transform(labels).astype(np.float64)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,labels,test_size=0.33,random_state=1) # Splitting Data set to train and test
```

```python
print('train_size = ',X_train.shape[0])
print('test_size = ',X_test.shape[0])
```

```python
## This function take number of neighbors and metric and weights along with train and test and give back the accuracy of classifier
def classiffier(k,p,X_train,X_test,y_train,y_test,weights):
    clf = KNN(n_neighbors=k,p=p,weights=weights)
    clf.fit(X_train,y_train)
    #print(f'The accuracy for the model with k ={k} p = {p} weights = {weights} is : {clf.score(X_test,y_test):.2}')
    return clf.score(X_test,y_test)
    
```

```python
# This function make a loop over the given data and draw 3 different charts (Manhattan,Euclidian and chebyshev) with different # neighbors
def accuracy_chart(X_train,X_test,y_train,y_test,weights):
    for p in(1,2,np.inf):
        scores = []
        plt.figure()
        for k in np.arange(1,50,3):
            score = classiffier(k,p,X_train,X_test,y_train,y_test,weights)
            scores.append(score)
        plt.plot( np.arange(1,50,3),np.array(scores))
        plt.title(f'metric[p={p}]  and weights = {weights}')
        plt.xlabel(f'The maximum is {max(scores)}')
    

        
```

## Now I have all pieces and I want to compare the accuracy result based on different # neighbors and different metrics

```python
accuracy_chart(X_train=X_train,X_test=X_test,y_train=y_train,y_test=y_test,weights='uniform')
```

### Maximum obtained on chebyshev and # neighbors = 7

### let's see what happens if we have different weights. I am using distance : weight points by the inverse of their distance. in this case, closer neighbors of a query point will have a greater influence than neighbors which are further away.

```python
accuracy_chart(X_train=X_train,X_test=X_test,y_train=y_train,y_test=y_test,weights='distance')
```

### This time results are better and the best result happens when I used Chebyshev and #neighbors = 31

```python
# Data points
X = np.array([[1,0], [1,1], [-2,4], [-2,-2],[1,3],[2,5],[-4,-5],[4,5],[2,-1]]) 
labels = [0,1,0,1,0,0,0,1,1]  # class_a = 0 ,class_b = 1
```

```python
new_data = np.array([[0,0],[-2,0],[5,5]])
```

```python
# lets draw the points and new points
plt.plot(X[:,0],X[:,1],'bo')
plt.plot(new_data[:,0],new_data[:,1],'rx')
plt.xlabel('x')
plt.ylabel('y')
```

```python
## This function will draw decision boundary for Data X with label y based on KNN alghorithm where it has k neighbors and
## use different metrics based on p(Minkowski metric) provided.
def draw_decision_boundary(k,p,X,y,new_data):
    # number of neighbors
    n_neighbors = k

    # distance metric 
    metric_p = p 


    # prepare data

    h = .02  

    # Create color maps
    cmap_light = ListedColormap(['#FFAAAA','#00AAD0'])
    cmap_bold = ListedColormap(['#FF0000','#00FFFF'])

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors,p=p)
    clf.fit(X, y)

    # calculate min, max and limits
    x_min, x_max = X[:, 0].min() - 2, X[:, 0].max() + 2
    y_min, y_max = X[:, 1].min() - 2, X[:, 1].max() + 2
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h))

    # predict class using data and kNN classifier
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    plt.plot(new_data[:,0],new_data[:,1],'x')   
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("2-Class classification (k = %i)" % (n_neighbors) )
    plt.show()
    predict = clf.predict(new_data)
    print("prediction to the points [0,0],[-2,0],[5,5] are ",predict[0],',',predict[1],',',predict[2],'respectively')
    return 
```
